```python
def open_file(ruta):
    csv_apertura= pd.read_csv(ruta, index_col=0)
    csv_apertura.reset_index(inplace = True)
    return csv_apertura
```

```python
# Funcion Exploración datos

def data_exploring(csv):
    forma = csv.shape
    print(f"La forma es {forma}")
    print("_______________")
    columnas = csv.columns
    print(f"Las columnas son {columnas}")
    print("_______________")
    nulos = ((csv.isna().sum()/csv.shape[0])*100).reset_index()
    print(f"Los nulos son:")
    display(nulos)
    print("_______________")
    duplicados = csv.duplicated().sum()
    print(f"Hay {duplicados} duplicados")
    print("_______________")
    tipo_dato = csv.dtypes.reset_index()
    print(f"Los datos son de tipo:")
    display(tipo_dato)
```

```python
for linea in df['Duration']:
    numeros = re.findall(r'\d+', linea)

    numeros_enteros = [int(numero) for numero in numeros]

    if len(numeros_enteros) == 2:
        resultado = float(numeros_enteros[0] + (numeros_enteros[1] / 100))
        print(resultado)
    elif len(numeros_enteros) == 1:
        resultado = int(numeros_enteros[0])
        print(resultado)
    else:
        print('Fallo en linea:', linea)
```

```python
def duraciones(linea):   
    numeros = re.findall(r'\d+', linea)

    numeros_enteros = [int(numero) for numero in numeros]

    if len(numeros_enteros) == 2:
        resultado = float((numeros_enteros[0] + (numeros_enteros[1] / 100)) * 60)
        return resultado
    
    elif len(numeros_enteros) == 1:
        resultado = int(numeros_enteros[0])
        return resultado
    else:
        return np.nan
```

```python
df2 = df.copy()
```

```python
df2['Duration'] = df2['Duration'].apply(duraciones)
```
