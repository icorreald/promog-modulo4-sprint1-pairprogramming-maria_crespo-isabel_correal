## Selenium Ranking

```python
from selenium import webdriver  
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.keys import Keys  
from selenium.webdriver.support.ui import Select  
import re
import pandas as pd
from selenium.webdriver.common.by import By

from time import sleep
```

```python
driver = webdriver.Chrome()

lista_urls = ['https://web.archive.org/web/20180328013733/https://myanimelist.net/topanime.php?type=tv', # 2018
              'https://web.archive.org/web/20190209000610/https://myanimelist.net/topanime.php?type=tv', # 2019
              'https://web.archive.org/web/20200110013254/https://myanimelist.net/topanime.php?type=tv', # 2020
              'https://web.archive.org/web/20210204020308/https://myanimelist.net/topanime.php?type=tv', # 2021
              'https://web.archive.org/web/20220204210002/https://myanimelist.net/topanime.php?type=tv', # 2022
              'https://web.archive.org/web/20230209181306/https://myanimelist.net/topanime.php?type=tv'] # 2023 
              
lista_total = []
patron = r'^.*?\n'

for url in lista_urls:
    lista_datos = []
    driver.get(url)
    sleep(1)
    try:
        driver.find_element('css selector', 'body > div.root > div > div.modal-wrapper > div > button').click()
    except:
        pass
    sleep(2)
    for child in range(2, 12):
        tupla_score = []

        # nombre del anime
        
        nombre_bruto = re.search(patron, driver.find_element('css selector', f'#content > div.pb12 > table > tbody > tr:nth-child({child}) > td.title.al.va-t.word-break').text)
        tupla_score.append(nombre_bruto.group(0).replace('\n', ''))

        # score del anime
        tupla_score.append(driver.find_element('css selector', f'#content > div.pb12 > table > tbody > tr:nth-child({child}) > td.score.ac.fs14').text)

        # añadir a lista global
        lista_datos.append(tuple(tupla_score))
    lista_total.append(lista_datos)
```

```python
lista_total
```

```python
definir_columns = [item for sublist in lista_total for item in sublist]
```

```python
ranking_anime = pd.DataFrame(definir_columns, columns=['title', 'score'])
```

```python
# añadir año

anyos = list(range(2018, 2024))

veces = [anyo for anyo in anyos for _ in range(10)]

ranking_anime["year"] = veces
```

```python
ranking_anime
```

```python
ranking_anime.to_csv("ranking_anime.csv")
```

# Selenium temporadas

```python
summer = {'2018':[],
            '2019':[],
            '2020':[],
            '2021':[],
            '2022':[],
            '2023':[]}
fall = {'2018':[],
            '2019':[],
            '2020':[],
            '2021':[],
            '2022':[],
            '2023':[]}
winter = {'2018':[],
            '2019':[],
            '2020':[],
            '2021':[],
            '2022':[],
            '2023':[]}
spring = {'2018':[],
            '2019':[],
            '2020':[],
            '2021':[],
            '2022':[],
            '2023':[]}
```

# Summer

```python
driver = webdriver.Chrome()
for year in range(2018, 2024):
    url = f'https://myanimelist.net/anime/season/{year}/summer'
    driver.get(url)
    sleep(1)
    for child in range(2, 11):
        datos_anime = []
        # nombre anime
        datos_anime.append(driver.find_element('css selector', f'#content > div.js-categories-seasonal > div:nth-child(1) > div:nth-child({child}) > div:nth-child(1) > div.title > div > h2 > a').text)
        datos_anime.append(driver.find_element('css selector', f'#content > div.js-categories-seasonal > div:nth-child(1) > div:nth-child({child}) > div.information > div.information-item.scormem > div').text)
        # datos_anime.append(driver.find_element('css selector', f'#content > div.js-categories-seasonal > div:nth-child(1) > div:nth-child({child}) > div.information > div.information-item.scormem > div > div.scormem-item.member').text)
        summer[f'{year}'].append(tuple(datos_anime))
```

```python
# Inicializar listas para almacenar los datos
animes = []
scores = []
years = []
members = []

# Iterar sobre el diccionario de datos
for year, anime_list in summer.items():
    for anime, score_info in anime_list:
        anime_name, score_and_members = anime, score_info.split('\n')
        score = score_and_members[0]
        member_info = score_and_members[1]
        members.append(member_info)
        animes.append(anime_name)
        scores.append(score)
        years.append(year)


# Crear DataFrame
df_summer = pd.DataFrame({'Anime': animes, 'Score': scores, 'Members': members,'Year': years })

# Visualizar DataFrame
print(df_summer)
```

```python
df_summer["season"] = "summer"

df_summer
```
